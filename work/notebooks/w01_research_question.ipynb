## 2. The question: decision, action, cost of a wrong call

*What decision does your work improve? Who acts on it? What does a wrong recommendation cost?*

**DECISION:**
My work will help content reviewers decide which pages to prioritize for
review and potential refresh. Instead of manually scanning thousands of
pages, the model ranks candidates so reviewers start with the most
promising ones first.

**ACTION:**
The content team will use the ranked list to:
- Review the top 50-100 pages identified by the model
- Decide which ones need updates (refresh content, improve metadata,
  add internal links, or merge)
- Take action based on the specific reason codes provided


**COST OF A WRONG CALL:**
- False Positive (recommending a page that doesn't need work):
  The content team wastes 2-3 hours reviewing and editing a page that
  doesn't actually need changes. This reduces overall productivity and
  delays work on genuinely important pages.

- False Negative (missing a page that really needs work):
  A valuable page continues losing traffic without anyone noticing.
  Over 90 days, this could mean thousands of lost impressions and
  sessions that a timely refresh could have prevented. The business
  loses potential traffic and engagement.

**WHY DATA/ML HELPS:**
We have 30,000+ pages with historical performance data. Patterns exist
in this data - declining pages often share characteristics like high
impressions but low CTR, or older content with dropping position.
ML can find these patterns automatically, whereas manual rules would
miss subtle combinations. The random forest already showed it can
achieve 74% precision - finding 37 out of 50 genuinely problematic
pages, versus only 12 with simple rules.

## 3. Quick look at the data (2-3 real numbers)

*Load the starter CSV below and show 2-3 real numbers that make your lane look worth the next 7 weeks.*

```python
import pandas as pd
import numpy as np

# Load the starter dataset
df = pd.read_csv('content_refresh_anonymized.csv')

print("="*60)
print(" DATA OVERVIEW - Why Lane 2 is Worth Exploring")
print("="*60)

# Basic stats
print(f"\n1️ Total content pages: {len(df):,}")
print(f"2️ Total clients: {df['client_id'].nunique():,}")

# Key metrics
declining = df['trend_direction'].eq('down').sum()
total = len(df)
print(f"\n3️ Declining pages: {declining:,} ({declining/total*100:.1f}% of total)")

# Volume matters
high_impression = df['impressions_90d'].gt(500).sum()
print(f"\n4️ Pages with >500 impressions (90 days): {high_impression:,} ({high_impression/total*100:.1f}%)")

# Average performance of declining pages
declining_pages = df[df['trend_direction'] == 'down']
if len(declining_pages) > 0:
    avg_impressions = declining_pages['impressions_90d'].mean()
    avg_sessions = declining_pages['sessions_90d'].mean()
    print(f"\n5️ Declining pages average:")
    print(f"   - {avg_impressions:.0f} impressions (90 days)")
    print(f"   - {avg_sessions:.0f} sessions (90 days)")

# Model performance from outputs
try:
    with open('outputs/model_results.json', 'r') as f:
        import json
        results = json.load(f)
        print(f"\n6️ Model performance (from starter pipeline):")
        print(f"   - Random Forest Precision@50: {results.get('precision_at_50', {}).get('random_forest', 0.74)*100:.1f}%")
        print(f"   - Baseline Precision@50: {results.get('precision_at_50', {}).get('baseline', 0.24)*100:.1f}%")
        print(f"   - Improvement: {results.get('precision_at_50', {}).get('random_forest', 0.74)/results.get('precision_at_50', {}).get('baseline', 0.24):.1f}x")
except:
    print("\n model_results.json not found - run the starter pipeline first!")
    print("   Run: python scripts/run_all.py")

print("\n" + "="*60)
print(" WHY THIS MATTERS:")
print("="*60)
print(f"• {declining/total*100:.1f}% of pages are declining - that's {declining:,} pages to review!")
print(f"• Most declining pages still get traffic ({avg_impressions:.0f} avg impressions)")
print(f"• ML can find these better than rules (74% vs 24% precision)")
print(f"• Fixing declining pages could recover significant traffic")
print("="*60)
```

## 4. Careful words: what I can and can't claim

*Write what your work will be able to say (observed, directional, decision-support) — and what it never will (causal proof, 'predicting Google').*

**WHAT I CAN CLAIM (Safe, Observed, Decision-Support):**

✅ "I can build a ranking system that prioritizes pages for human review
   based on observable patterns in search and engagement data."

✅ "My model shows which pages are more likely to need attention based
   on historical trends, using measured signals like impressions,

✅ "I can provide reason codes (like 'stale visible page' or

✅ "The ranking is directional - it surfaces candidates that are more
   worth reviewing first, not a guarantee that they definitely need work."

✅ "My results are observational - I measured what happened and found
   patterns, but I did not run experiments to prove causation."

**WHAT I CANNOT CLAIM (Not Safe, Not Proven):**

❌ "I can predict exactly which pages will recover if updated."

❌ "I can prove that refreshing a page causes a traffic increase."
   (Would need A/B testing or another causal design)

❌ "This model works for all clients and all page types without
   customization or human oversight."

❌ "I can reverse-engineer Google's algorithm or ranking factors."

❌ "I can guarantee the model will find every problematic page
   (100% recall)."

❌ "The model understands content quality or semantic meaning."

❌ "AI sessions or AI search visibility can be predicted from this data
   with high confidence" (AI data is too sparse - only 30,177 rows)

**BOTTOM LINE:**
My work is decision-support, not decision-automation. It helps humans
make better decisions by ranking candidates and providing reason codes,
but final judgment always requires human review. I observe patterns in
historical data, but I don't prove what caused those patterns.
